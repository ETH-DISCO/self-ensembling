- mean accuracy, when grouped by: dataset, train_epochs, train_hcaptcha_ratio, train_opacity

````r
deps <- c("ISLR", "ndjson", "ggplot2", "dplyr", "caret", "tidyr", "gridExtra", "data.table", "repr")
for (p in deps) {
    if (!requireNamespace(p, quietly = TRUE)) {
        install.packages(p, repos = "https://cran.rstudio.com")
    }
    library(p, character.only = TRUE)
}
options(scipen=999)
options(repr.plot.width=25, repr.plot.height=25)

df <- ndjson::stream_in("./resnet.jsonl")


# group by columns: dataset, train_epochs, train_hcaptcha_ratio, train_opacity
# get mean of columns: acc, acc_0, acc_2, acc_4, acc_8, acc_16, acc_32, acc_64, acc_128
df <- df %>%
    group_by(dataset, train_epochs, train_hcaptcha_ratio, train_opacity) %>%
    summarise(
        acc = mean(acc),
        acc_0 = mean(acc_0),
        acc_2 = mean(acc_2),
        acc_4 = mean(acc_4),
        acc_8 = mean(acc_8),
        acc_16 = mean(acc_16),
        acc_32 = mean(acc_32),
        acc_64 = mean(acc_64),
        acc_128 = mean(acc_128)
    )

df
````
